```python
prevalent_colo = ['Food & Snack', 'Grocery & Market', 'History, Museum & Arts', 'Hotel', 'Indoor Entertainment', 'Music', 'Office', 'Outdoor Entertainment', 'Place for Socializing', 'Place for Sports', 'Residence', 'Restaurant', 'School', 'Tobacco & Alcohol', 'Transportation']
cities = ['amsterdam', 'barcelona', 'berlin', 'helsinki', 'london', 'moscow', 'newyork', 'paris', 'prague', 'rome', 'stockholm']
```

```python

dictionary = {}
for city in cities:
    data_path = "./big4sqColoMpi/{}_colo.csv".format(city)
    data_colo = pd.read_csv(data_path)
    dd = dict()
    print(city)
    for guo_index in prevalent_colo:
        guo = np.array(data_colo[guo_index]).reshape((-1,1))
        guo = guo[1:]-guo[:-1]
        causes = []
        for colo in prevalent_colo:
            colo = str(colo)
            if colo!=guo_index:
                yin = np.array(data_colo[colo]).reshape((-1,1))
                yin = yin[1:]-yin[:-1]
                df = np.concatenate((guo,yin),axis=1)
                lags = select_time_lags(yin)
#                 print(lags)
                results = grangercausalitytests(df, maxlag=[lags],verbose = False)
                if (results[lags])[0]['ssr_ftest'][1]<=0.15:
                    causes.append(colo)
        #print("{}, {}".format(guo_index,causes))
        dd[guo_index] = causes
        
    dictionary[city] = dd
#     print(dd)
        
    for key in dd:
        causes = dd[key]
        y = str(key)
        dff = pd.DataFrame()
        for pattern in causes:
            x = np.array(data_colo[pattern])
            x = x[1:]-x[:-1]
            dff[pattern] = x
        t = np.array(data_colo[y])
        t = t[1:] - t[:-1]
        dff[y] = t
        dff.to_csv("./city_granger/{}_{}.csv".format(city,y))
```
