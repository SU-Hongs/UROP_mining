```python
import pandas as pd

data=pd.read_csv('dataset_tsmc2014.num_colo.csv')
data=data.drop(data.columns[[0]],axis=1)
data.head()
```

```python
from causalnex.structure.notears import from_pandas

struct_data=data.copy()
sm=from_pandas(struct_data)
sm
```

```python
from causalnex.plots import plot_structure
import networkx as nx
import matplotlib.pyplot as plt

sm_thres=sm.copy()
sm_thres.remove_edges_below_threshold(1.3)


viz=plot_structure(sm_thres,prog='neato',
    graph_attributes={'bgcolor':'white','scale':'2'},all_node_attributes={'fontcolor':'black','fontsize':12},
    all_edge_attributes={'color':'#0b4787'})

viz.draw(format='png',path='./causalnet_output.png')

# nx.draw(G, cmap = plt.get_cmap('jet'), node_color = values)
# plt.show()
```
