```python
import random
import numpy as np
import matplotlib.pyplot as plt

eps=[2*random.random()-1 for i in range(1005)]
xs=eps.copy()
for i in range(1,1005): xs[i]+=xs[i-1]
ys=[2*xs[i]+random.random()*0.02-0.01 for i in range(5,1005)]
xs=xs[:1000] 
corr=np.corrcoef(xs,ys)[0][1]
plt.scatter(xs,ys)
plt.title('corr='+str(corr))
plt.show()
```
